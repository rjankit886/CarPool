```python
from tkinter import *
from PIL import Image,ImageTk
from tkinter.ttk import Combobox
from tkinter import messagebox
from tkinter.scrolledtext import ScrolledText
from tkinter import filedialog
import os
import sqlite3
import shutil
```

```python
try: 
    con = sqlite3.connect(database="carpool.sqlite")
    cursor = con.cursor()
    table1 = "create table users(username text primary key,password text,mobile text,email text)"
    table2 = "create table cars(username text,src text,dest text,regno text primary key,model text,seats text,fuel text,gtime,rtime)"
    cursor.execute(table1)
    cursor.execute(table2)
    con.commit()
    con.close()
    print("Tables Created")
except Exception as e:
    print(e)
```

```python
win = Tk()
win.state("zoomed")
win.configure(bg="powder blue")
lbl_title = Label(win,text = "Car Pooling",font = ('',80,'bold','underline'),bg="powder blue",fg="red")
lbl_title.pack()

img = Image.open("Images/cp4.jpg").resize((400,150))
imgtk=ImageTk.PhotoImage(img,master=win)
lbl_img=Label(win,image=imgtk)
lbl_img.place(relx=.0,rely=0)

img2 = Image.open("Images/cp2.jpg").resize((400,150))
imgtk2=ImageTk.PhotoImage(img2,master=win)
lbl_img2=Label(win,image=imgtk2)
lbl_img2.place(relx=.76,rely=0)

img3 = Image.open("Images/login.png").resize((140,40))
imgtk3=ImageTk.PhotoImage(img3)

img4 = Image.open("Images/reset.png").resize((140,40))
imgtk4=ImageTk.PhotoImage(img4)

img5 = Image.open("Images/logout.png").resize((140,40))
imgtk5=ImageTk.PhotoImage(img5)

img6 = Image.open("Images/register.png").resize((140,40))
imgtk6=ImageTk.PhotoImage(img6)

img7 = Image.open("Images/newuser.png").resize((200,40))
imgtk7=ImageTk.PhotoImage(img7)

img8 = Image.open("Images/back.png").resize((100,40))
imgtk8=ImageTk.PhotoImage(img8)

img9 = Image.open("Images/search.png").resize((140,40))
imgtk9=ImageTk.PhotoImage(img9)

img10 = Image.open("Images/submit.png").resize((140,40))
imgtk10=ImageTk.PhotoImage(img10)

img11 = Image.open("Images/update.png").resize((140,40))
imgtk11=ImageTk.PhotoImage(img11)

img12 = Image.open("Images/profilepic.jpg").resize((135,140))
imgtk12=ImageTk.PhotoImage(img12)

def home_frame():
    frm = Frame(win,highlightthickness=3,highlightbackground="brown")
    frm.place(x=400,y=250,relwidth=.5,relheight=.45)
    
    def newuser(event):
        frm.destroy()
        newuser_frame()
        
    def login(event):
        global loginuser_row
        u=e_user.get()
        p=e_pass.get()
        
        if(len(u)==0 or len(p)==0):
            messagebox.showwarning('Validation',"Please fill all fields!!")
        else:
            try: 
                con = sqlite3.connect(database="carpool.sqlite")
                cursor = con.cursor()
                cursor.execute("select * from users where username=? and password=?",(u,p))
                loginuser_row = cursor.fetchone()
                con.close()

                if(loginuser_row==None):
                    messagebox.showerror("Login","Invalid Username/Password")
                else:
                    frm.destroy()
                    login_frame()

            except Exception as e:
                messagebox.showerror("Login",str(e))

        
    def reset(event):
        e_user.delete(0,"end")
        e_pass.delete(0,"end")
        e_user.focus()
    
    lbl_user = Label(frm,text="Username",font=('',20,'bold'))
    lbl_pass = Label(frm,text="Password",font=('',20,'bold'))
    e_user = Entry(frm,font=('',20,'bold'),bd=5)
    e_user.focus()
    e_pass = Entry(frm,font=('',20,'bold'),bd=5,show="*")
    
    lbl_user.place(x=150,y=50)
    e_user.place(x=320,y=50)
    lbl_pass.place(x=150,y=130)
    e_pass.place(x=320,y=130)
    
    btn_login = Label(frm,image=imgtk3)
    btn_login.place(x=240,y=210)
    btn_login.bind("<Button>",login)
    
    btn_reset = Label(frm,image=imgtk4)
    btn_reset.place(x=390,y=210)
    btn_reset.bind("<Button>",reset)
    
    btn_register = Label(frm,image=imgtk7)
    btn_register.place(x=280,y=280)
    btn_register.bind("<Button>",newuser)
    
def newuser_frame():
    frm = Frame(win,highlightthickness=3,highlightbackground="brown")
    frm.place(x=400,y=250,relwidth=.5,relheight=.55)
    
    def reg_todb(event):
        u=e_user.get()
        p=e_pass.get()
        m=e_mob.get()
        e=e_email.get()
        
        if(len(u)==0 or len(p)==0 or len(m)==0 or len(e)==0):
            messagebox.showwarning('Validation',"Please fill all fields!!")
        else:
            try: 
                con = sqlite3.connect(database="carpool.sqlite")
                cursor = con.cursor()
                cursor.execute("insert into users values(?,?,?,?)",(u,p,m,e))
                con.commit()
                con.close()
                messagebox.showinfo("Registration","Registration Done!!")
                frm.destroy()
                home_frame()
            except Exception as e:
                messagebox.showerror("Registration","Username already exist")
                
    def reset(event):
        e_user.delete(0,"end")
        e_pass.delete(0,"end")
        e_mob.delete(0,"end")
        e_email.delete(0,"end")
        e_user.focus()

    lbl_user = Label(frm,text="Username",font=('',20,'bold'))
    lbl_pass = Label(frm,text="Password",font=('',20,'bold'))
    lbl_mob = Label(frm,text="Mobile No.",font=('',20,'bold'))
    lbl_email = Label(frm,text="Email",font=('',20,'bold'))
    
    e_user = Entry(frm,font=('',20,'bold'),bd=5)
    e_user.focus()
    e_pass = Entry(frm,font=('',20,'bold'),bd=5,show="*")
    e_mob = Entry(frm,font=('',20,'bold'),bd=5,)
    e_email = Entry(frm,font=('',20,'bold'),bd=5,)

    
    lbl_user.place(x=150,y=50)
    e_user.place(x=320,y=50)
    lbl_pass.place(x=150,y=130)
    e_pass.place(x=320,y=130)
    lbl_mob.place(x=150,y=210)
    e_mob.place(x=320,y=210)
    lbl_email.place(x=150,y=290)
    e_email.place(x=320,y=290)
    
    btn_register = Label(frm,image=imgtk6)
    btn_register.place(x=250,y=370)
    btn_register.bind("<Button>",reg_todb)
    
    btn_reset = Label(frm,image=imgtk4)
    btn_reset.place(x=400,y=370)
    btn_reset.bind("<Button>",reset)
    
def login_frame():
    frm = Frame(win,highlightthickness=3,highlightbackground="brown")
    frm.place(x=300,y=220,relwidth=.65,relheight=.55)
    
    def logout(event):
        frm.destroy()
        home_frame()
        
    def search():
        frm.destroy()
        searchpool_frame()
        
    def create():
        frm.destroy()
        createpool_frame()
    
    def picture():
        img_path = filedialog.askopenfilename()
        shutil.copy(img_path,f"Images/{loginuser_row[0]}.png")
        img = Image.open(f"Images/{loginuser_row[0]}.png").resize((135,140))
        imgtk=ImageTk.PhotoImage(img)
        lbl_profilepic['image'] = imgtk
        lbl_profilepic.image = imgtk
        
    def update():
        frm.destroy()
        updateprofile_frame()
        
    lbl_wel = Label(frm,text=f"Welcome {loginuser_row[0]}",font=('',10,'bold'),fg="green")
    lbl_wel.place(x=0,y=0)
    
    lbl_profilepic = Label(frm,image=imgtk12)
    lbl_profilepic.place(x=0,y=40)
        
    if(os.path.exists(f"Images/{loginuser_row[0]}.png")):
        img = Image.open(f"Images/{loginuser_row[0]}.png").resize((135,140))
        imgtk=ImageTk.PhotoImage(img)
        lbl_profilepic['image'] = imgtk
        lbl_profilepic.image = imgtk
    
    btn_logout = Label(frm,image=imgtk5)
    btn_logout.place(relx=.85,rely=0)
    btn_logout.bind("<Button>",logout)
    
    btn_search = Button(frm,text="Search Pool",font=('',10,'bold'),bd=5,width=20,command=search)
    btn_search.place(relx=.4,rely=.1)
    
    btn_create = Button(frm,text="Create Pool",font=('',10,'bold'),bd=5,width=20,command=create)
    btn_create.place(relx=.4,rely=.3)
    
    btn_update = Button(frm,text="Update Profile",font=('',10,'bold'),bd=5,width=20,command=update)
    btn_update.place(relx=.4,rely=.5)
    
    btn_profilepic = Button(frm,text="Profile Picture",font=('',10,'bold'),bd=5,width=20,command=picture)
    btn_profilepic.place(relx=.4,rely=.7)
    
def searchpool_frame():
    frm = Frame(win,highlightthickness=3,highlightbackground="brown")
    frm.place(x=300,y=220,relwidth=.65,relheight=.55)
    
    def logout(event):
        frm.destroy()
        home_frame()
        
    def back(event):
        frm.destroy()
        login_frame()
        
    def searchpool_db(event):
        src = e_src.get().upper()
        dest = e_dest.get().upper()
        
        if(len(src)==0 or len(dest)==0):
            messagebox.showwarning('Validation',"Please fill all fields!!")
        else:
            try: 
                con = sqlite3.connect(database="carpool.sqlite")
                cursor = con.cursor()
                cursor.execute("select * from cars where src=? and dest=?",(src,dest))
                pools = cursor.fetchall()
                print(pools)
                con.close()

                if(len(pools)==0):
                    messagebox.showerror("Pool","Pool does not exits")
                else:
                    con = sqlite3.connect(database="carpool.sqlite")
                    cursor = con.cursor()
                    st = ScrolledText(frm,width=115,height=18,font=('Arial',10,'bold','underline'))
                    st.place(x=90,y=130)
                    st.insert("end","Username\t\tCar Model\t\tReg. No.\t\tSeats\tFuel Type\t\tMobile\t\tEmail\t\tgtime\t\trtime\n")
                    for pool in pools:
                        cursor.execute("select mobile,email from users where username=?",(pool[0],))
                        m,e = cursor.fetchone()
                        st.insert("end",f"\n{pool[0]}\t\t{pool[4]}\t\t{pool[3]}\t\t{pool[5]}\t{pool[6]}\t\t{m}\t\t{e}\t\t{pool[7]}\t\t{pool[8]}","clr")
                        st.tag_config("clr",font=('',10),foreground='red')
                    st.configure(state="disabled")
                    con.close()

            except Exception as e:
                messagebox.showerror("Pool",str(e))
        
    lbl_wel = Label(frm,text=f"Welcome {loginuser_row[0]}",font=('',15,'bold'),fg="green")
    lbl_wel.place(x=0,y=0)
    
    btn_logout = Label(frm,image=imgtk5)
    btn_logout.place(relx=.85,rely=0)
    btn_logout.bind("<Button>",logout)
    
    btn_back = Label(frm,image=imgtk8)
    btn_back.place(relx=0,rely=.1)
    btn_back.bind("<Button>",back)
    
    lbl_src = Label(frm,text="Source",font=('',15,'bold'),fg="blue")
    lbl_src.place(relx=.2,rely=.1)
    e_src = Entry(frm,font=('',15),bd=3,width=10)
    e_src.focus()
    e_src.place(relx=.28,rely=.1)
    
    lbl_dest = Label(frm,text="Destination",font=('',15,'bold'),fg="blue")
    lbl_dest.place(relx=.46,rely=.1)
    e_dest = Entry(frm,font=('',15),bd=3,width=10)
    e_dest.place(relx=.58,rely=.1)
    
    btn_search = Label(frm,image=imgtk9)
    btn_search.place(relx=.4,rely=.2)
    btn_search.bind("<Button>",searchpool_db)

def createpool_frame():
    frm = Frame(win,highlightthickness=3,highlightbackground="brown")
    frm.place(x=300,y=220,relwidth=.65,relheight=.55)
    
    def logout(event):
        frm.destroy()
        home_frame()
        
    def back(event):
        frm.destroy()
        login_frame()
        
    def createpool_db(event):
        src = e_src.get()
        dest = e_dest.get()
        model = e_model.get()
        seats = cb_seats.get()
        regno = e_regno.get().replace(" ","")
        fuel = cb_fuel.get()
        gtime = e_gtime.get()
        rtime = e_rtime.get()
        
        if(len(src)==0 or len(dest)==0 or len(model)==0 or len(seats)==0 or len(regno)==0 or len(fuel)==0 or len(gtime)==0 or len(rtime)==0):
            messagebox.showwarning('Validation',"Please fill all fields!!")
        else:
            try: 
                con = sqlite3.connect(database="carpool.sqlite")
                cursor = con.cursor()
                cursor.execute("insert into cars values(?,?,?,?,?,?,?,?,?)",(loginuser_row[0],src.upper(),dest.upper(),regno.upper(),model.upper(),seats,fuel.upper(),gtime,rtime)
                con.commit()
                con.close()
                messagebox.showinfo("Create Pool","Pool created Successfully!!")
                frm.destroy()
                login_frame()
            except Exception as e:
                messagebox.showerror("Create Pool","This car already exist in pool")

        
    def reset(event):
        e_src.delete(0,"end")
        e_dest.delete(0,"end")
        e_model.delete(0,"end")
        e_regno.delete(0,"end") 
        cb_seats.delete(0,"end")
        cb_fuel.delete(0,"end")
        e_gtime.delete(0,"end")
        e_rtime.delete(0,"end") 
        e_src.focus()
        
    lbl_wel = Label(frm,text=f"Welcome {loginuser_row[0]}",font=('',15,'bold'),fg="green")
    lbl_wel.place(x=0,y=0)
    
    btn_logout = Label(frm,image=imgtk5)
    btn_logout.place(relx=.85,rely=0)
    btn_logout.bind("<Button>",logout)
    
    btn_back = Label(frm,image=imgtk8)
    btn_back.place(relx=0,rely=.1)
    btn_back.bind("<Button>",back)
    
    lbl_src = Label(frm,text="Source",font=('',15,'bold'),fg="blue")
    lbl_src.place(relx=.2,rely=.15)
    e_src = Entry(frm,font=('',15),bd=3,width=12)
    e_src.focus()
    e_src.place(relx=.32,rely=.15)
    
    lbl_dest = Label(frm,text="Destination",font=('',15,'bold'),fg="blue")
    lbl_dest.place(relx=.5,rely=.15)
    e_dest = Entry(frm,font=('',15),bd=3,width=12)
    e_dest.place(relx=.62,rely=.15)
    
    lbl_model = Label(frm,text="Car Model",font=('',15,'bold'),fg="blue")
    lbl_model.place(relx=.2,rely=.3)
    e_model = Entry(frm,font=('',15),bd=3,width=12)
    e_model.place(relx=.32,rely=.3)
    
    lbl_seats = Label(frm,text="Seats",font=('',15,'bold'),fg="blue")
    lbl_seats.place(relx=.5,rely=.3)
    cb_seats = Combobox(frm,values=[2,3,4,5,6,7],font=('',10,'bold'))
    cb_seats.current(2)
    cb_seats.place(relx=.62,rely=.3)
    
    lbl_regno = Label(frm,text="Car Reg.No.",font=('',15,'bold'),fg="blue")
    lbl_regno.place(relx=.2,rely=.45)
    e_regno = Entry(frm,font=('',15),bd=3,width=12)
    e_regno.place(relx=.32,rely=.45)
    
    lbl_fuel = Label(frm,text="Fuel Type",font=('',15,'bold'),fg="blue")
    lbl_fuel.place(relx=.5,rely=.45)
    cb_fuel = Combobox(frm,values=['Diesel','Petrol','Petrol+CNG','Petrol+LPG','Electronic'],font=('',10,'bold'))
    cb_fuel.current(2)
    cb_fuel.place(relx=.62,rely=.45)
    
    lbl_gtime = Label(frm,text="Going Time",font=('',15,'bold'),fg="blue")
    lbl_gtime.place(relx=.2,rely=.6)
    e_gtime = Entry(frm,font=('',15),bd=3,width=12)
    e_gtime.place(relx=.32,rely=.6)
    
    lbl_rtime = Label(frm,text="Ret. Time",font=('',15,'bold'),fg="blue")
    lbl_rtime.place(relx=.5,rely=.6)
    e_rtime = Entry(frm,font=('',15),bd=3,width=12)
    e_rtime.place(relx=.62,rely=.6)
    
    btn_submit = Label(frm,image=imgtk10)
    btn_submit.place(relx=.35,rely=.75)
    btn_submit.bind("<Button>",createpool_db)

    btn_reset = Label(frm,image=imgtk4)
    btn_reset.place(relx=.5,rely=.75)
    btn_reset.bind("<Button>",reset)

def updateprofile_frame():
    frm = Frame(win,highlightthickness=3,highlightbackground="brown")
    frm.place(x=400,y=250,relwidth=.5,relheight=.55)
    
    def logout(event):
        frm.destroy()
        home_frame()
        
    def back(event):
        frm.destroy()
        login_frame()
        
    def update_todb(event):
        u = e_user.get()
        p = e_pass.get()
        m = e_mob.get()
        e = e_email.get()
        
        con = sqlite3.connect(database="carpool.sqlite")
        cursor = con.cursor()
        cursor.execute("update users set password=?,mobile=?,email=? where username=?",(p,m,e,u))
        con.commit()
        con.close()
        messagebox.showinfo("Update profile","Record Updated")
        frm.destroy()
        home_frame()
        
    def reset(event):
        e_user.delete(0,"end")
        e_pass.delete(0,"end")
        e_mob.delete(0,"end")
        e_email.delete(0,"end")
        e_user.focus()
    
    lbl_user = Label(frm,text="Username",font=('',20,'bold'))
    lbl_pass = Label(frm,text="Password",font=('',20,'bold'))
    lbl_mob = Label(frm,text="Mobile No.",font=('',20,'bold'))
    lbl_email = Label(frm,text="Email",font=('',20,'bold'))
    
    e_user = Entry(frm,font=('',20,'bold'),bd=5)
    e_user.focus()
    e_pass = Entry(frm,font=('',20,'bold'),bd=5,show="*")
    e_mob = Entry(frm,font=('',20,'bold'),bd=5,)
    e_email = Entry(frm,font=('',20,'bold'),bd=5,)

    
    lbl_user.place(x=150,y=50)
    e_user.place(x=320,y=50)
    lbl_pass.place(x=150,y=130)
    e_pass.place(x=320,y=130)
    lbl_mob.place(x=150,y=210)
    e_mob.place(x=320,y=210)
    lbl_email.place(x=150,y=290)
    e_email.place(x=320,y=290)
    
    btn_update = Label(frm,image=imgtk11)
    btn_update.place(x=250,y=370)
    btn_update.bind("<Button>",update_todb)
    
    btn_reset = Label(frm,image=imgtk4)
    btn_reset.place(x=400,y=370)
    btn_reset.bind("<Button>",reset)

    lbl_wel = Label(frm,text=f"Welcome {loginuser_row[0]}",font=('',15,'bold'),fg="green")
    lbl_wel.place(x=0,y=0)
    
    btn_logout = Label(frm,image=imgtk5)
    btn_logout.place(relx=.81,rely=0)
    btn_logout.bind("<Button>",logout)
    
    btn_back = Label(frm,image=imgtk8)
    btn_back.place(relx=0,rely=.1)
    btn_back.bind("<Button>",back)
    
    con = sqlite3.connect(database="carpool.sqlite")
    cursor = con.cursor()
    cursor.execute("select mobile,email,password from users where username=?",(loginuser_row[0],))
    row = cursor.fetchone()
    e_user.configure(state="normal")
    e_user.delete(0,"end")
    e_user.insert(0,loginuser_row[0])
    e_user.configure(state="disabled")
    
    e_pass.delete(0,"end")
    e_pass.insert(0,row[2])
    
    e_mob.delete(0,"end")
    e_mob.insert(0,row[0])
    
    e_email.delete(0,"end")
    e_email.insert(0,row[1])

    con.close()

home_frame()
win.mainloop()
```
